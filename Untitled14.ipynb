```python
import torch
from transformers import MarianMTModel, MarianTokenizer
import sacrebleu

# 🔹 Small manual dataset (no heavy loading)
english_sentences = [
    "Hello, how are you?",
    "This is a test sentence.",
    "Machine translation is interesting.",
    "Keep calm and relax.",
    "I like learning new things."
]

french_references = [
    ["Bonjour, comment ça va ?"],
    ["Ceci est une phrase de test."],
    ["La traduction automatique est intéressante."],
    ["Reste calme et détends-toi."],
    ["J'aime apprendre de nouvelles choses."]
]

# 🔹 Load model
model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# 🔹 CPU only
device = torch.device("cpu")
model = model.to(device)

# 🔹 Tokenize
inputs = tokenizer(
    english_sentences,
    return_tensors="pt",
    padding=True,
    truncation=True,
    max_length=50
)
inputs = {k: v.to(device) for k, v in inputs.items()}

# 🔹 Generate
with torch.no_grad():
    outputs = model.generate(**inputs, max_length=50)

translations = tokenizer.batch_decode(outputs, skip_special_tokens=True)

# 🔹 Print results (altered format)
print("\n=== TRANSLATION OUTPUT ===\n")

for i in range(len(english_sentences)):
    print(f"[{i+1}] EN: {english_sentences[i]}")
    print(f"    FR(pred): {translations[i]}")
    print(f"    FR(ref) : {french_references[i][0]}\n")

# 🔹 BLEU
bleu = sacrebleu.corpus_bleu(translations, french_references)
print("Final BLEU Score:", round(bleu.score, 2))
```
