```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from keras_facenet import FaceNet
from google.colab import files

# Initialize FaceNet
embedder = FaceNet()

# Haar Cascade
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# ---------------------------
# Extract Faces
# ---------------------------
def extract_faces(img):
    if img is None:
        return []

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)

    face_list = []

    for (x, y, w, h) in faces:
        face = img[y:y+h, x:x+w]

        if face.size == 0:
            continue

        face = cv2.resize(face, (160,160))
        face_list.append((face, (x, y, w, h)))

    return face_list

# ---------------------------
# Get Embedding (Normalized)
# ---------------------------
def get_embedding(face):
    face = face.astype('float32')
    face = np.expand_dims(face, axis=0)
    emb = embedder.embeddings(face)[0]
    return emb / np.linalg.norm(emb)

# ---------------------------
# Upload Images
# ---------------------------
print("📂 Upload TWO images:")
print("1️⃣ Known person (prefer single clear face)")
print("2️⃣ Test image (can be group)")

uploaded = files.upload()
files_list = list(uploaded.keys())

if len(files_list) < 2:
    print("❌ Please upload at least 2 images")
else:
    known_image = files_list[0]
    test_image = files_list[1]

    # ---------------------------
    # Process Known Image
    # ---------------------------
    known_embedding = None
    known_name = "Person1"

    img_known = cv2.imread(known_image)

    faces = extract_faces(img_known)

    if len(faces) == 0:
        print("❌ No face detected in known image")
    else:
        # pick largest face
        face = max(faces, key=lambda x: x[1][2]*x[1][3])[0]
        known_embedding = get_embedding(face)
        print("✅ Known face loaded")

    # ---------------------------
    # Process Test Image
    # ---------------------------
    img_test = cv2.imread(test_image)

    if img_test is None:
        print("❌ Test image not loaded")

    elif known_embedding is None:
        print("❌ Cannot proceed without known face")

    else:
        rgb = cv2.cvtColor(img_test, cv2.COLOR_BGR2RGB)
        faces = extract_faces(img_test)

        if len(faces) == 0:
            print("❌ No face detected in test image")

        else:
            for face, (x, y, w, h) in faces:
                emb = get_embedding(face)

                dist = np.linalg.norm(known_embedding - emb)

                # stricter threshold
                if dist < 0.7:
                    identity = known_name
                else:
                    identity = "Unknown"

                print(f"Distance: {dist:.2f} → {identity}")

                # Draw box
                cv2.rectangle(rgb, (x, y), (x+w, y+h), (0,255,0), 2)
                cv2.putText(rgb, identity, (x, y-10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,255,0), 2)

            plt.imshow(rgb)
            plt.title("Face Recognition Result")
            plt.axis('off')
            plt.show()
```
